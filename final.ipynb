```python
import json


import pandas as pd


#transactions_final

with open('transactions_final.json') as f:
    transactions = json.load(f)

list_trans_id = []

dict_transactions = {}

for key, transaction in transactions['transactions'].items():
    list_trans_id.append(key)
    for k, v in transaction.items():
        if not dict_transactions.get(k):
            dict_transactions[k] = []
            dict_transactions[k].append(v)
        else:
            dict_transactions[k].append(v)

dict_transactions['transaction_id'] = list_trans_id
```

```python
df_new_pattern = pd.DataFrame(dict_transactions) # data frame для нового паттерна
df_new_pattern.head(4)
```

```python
# Удаляем ненужные столбцы
df_new_pattern.pop('account_valid_to')
df_new_pattern.pop('date_of_birth')
df_new_pattern.pop('city')
df_new_pattern.pop('oper_type')
df_new_pattern.pop('amount')
df_new_pattern.pop('oper_result')
df_new_pattern.pop('terminal')
df_new_pattern.pop('terminal_type')
```

```python
# Собираем все паспорта
passports = []
for i in transactions["transactions"].keys():
    fio = transactions["transactions"][i]['passport']
    if fio not in passports:
        passports.append(fio)
```

```python
# Собираем все транзакции для каждого пасспорта
trans_np = {}
for passport in passports:
    tran = df_new_pattern.loc[df_new_pattern['passport'] == passport]
    if not(trans_np.get(passport)):
        trans_np[passport] = [tran]
    else:
        trans_np[passport].append(tran)


```

```python
# Находим кол-во транзакций и их время

ans = [0 for i in range(22)]
date_passport = {}
date_passport_ids = {}
for passort in trans_np.keys():

    a = trans_np[passort][0]
    a = pd.DataFrame(a)
    date_passport[passort] = a['date'].tolist()
    date_passport_ids[passort] = a['transaction_id'].tolist()
    count_trans = len(date_passport[passort])

    ans[count_trans] += 1

for i in range(len(ans)):
    print(i,':',ans[i])
# Кол-во транзакций от каждого пасспорта, Больше 20 транзакций никто не совершал
```
